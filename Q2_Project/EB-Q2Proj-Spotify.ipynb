```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

### Original Data Set

```python
df = pd.read_csv("spotify_songs.csv")
df.head()        
```

```python
df['playlist_genre'].unique()
```

```python
plt.scatter(df["track_popularity"], df["duration_ms"],color = "C7", alpha=0.03)

plt.title('Populatiry vs Duration')
plt.xlabel('Track popularity')
plt.ylabel('Duration in ms')


```

```python
stats.pearsonr(df["track_popularity"], df["duration_ms"])
```

## The negative value represents a negative correlation between popularity and duration indicating that the shorter songs tend to be more popular. The p-value shows that there is very high confidence.

### My Two Figures

```python
fig, (ax1, ax2) = plt.subplots(2,1)
ax1.hist2d(df["track_popularity"], df["tempo"])
ax2.hist2d(df["track_popularity"], df["valence"])

plt.title("Popularity vs Tempo and Popularity vs Valence")
ax1.set_ylabel("Tempo")
ax1.set_xlabel("Track Popularity")
ax2.set_ylabel("Valence")
ax2.set_xlabel("Track Popularity")

fig.tight_layout()
```
