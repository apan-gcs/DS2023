### [2] Scatter Plot
Below is a simple scatter plot that shows the representation between when a house was built to its price.

```python
fig, ax = plt.subplots()

ax.scatter(df["YearBuilt"], df["Price"], alpha=0.1)
ax.set_xlim(1960, 2000)
ax.set_title("Year Built vs Price")
ax.set_ylabel("Price")
ax.set_xlabel("Year")
```

### [3] Comparing price and squarefeet.
Below is two plots, one plottting square feet vs the year it was built, and the other plotting price vs year built. In both graphs, their plotted lines are very similar. With this information, you can make an inference on how the sqaure footage of a house has a driect relationshop with the price of the house. 

```python
fig3, (ax31, ax32) = plt.subplots(2)
fig3.tight_layout()
fig3.suptitle("The Avg. Price of a Home Compared to When it Was Built And The Avg. Square Footage of a Home Compared to When it Was Built ")
ax31.plot(df.groupby("YearBuilt").mean()["Price"])
ax32.plot(df.groupby("YearBuilt").mean()["SquareFeet"])
ax31.set_xlabel("Year Built")
ax31.set_ylabel("Mean Price")

ax32.set_ylabel("Square Feet")
ax32.set_xlabel("Year Built")
#df.plot(kind='line',x="YearBuilt",y="Price",ax=ax)
#df.plot(kind='line',x="Price",y="YearBuilt", color='red', ax=ax)

#plt.show()
```

### [4] Price of a home per neighborhood
Below is a histogram representing the average price of a home in each different type of neighborhood. One inference that could be made about this data set from the histogram is how unbalanced the datset is. If you were to take the average price of a urban homes, I am sure it would be larger than the average price of all rural homes. However, with this dataset, all avergae price values are the same, representing how clean this data is.

```python
# Fig 4. Plot bar graph for various neighborhoods
Neighborhoods = ['Rural', 'Urban', 'Surburban']

fig4, ax4 = plt.subplots()

ax4.bar(Neighborhoods, df.groupby("Neighborhood").mean()["Price"])

# Format / Rotate labels
labels = ['Rural', 'Urban','Surburban']
ax4.set_xticklabels(labels, rotation=35, ha='right')

ax4.set_title("Avg. Price of home per neighborhood type")
ax4.set_ylabel("Average house Price")

ax4.set_ylim(200000, 240000)

plt.show()



```

### [5] Correlation
This code calculates the Pearson correlation coefficient and a p-value to test for a linear relationship between the "Price" and "SquareFeet". It has been determined that due to the p-value being zero, the null hypothesis can be rejected

```python
x = df["Price"]
y = df["SquareFeet"]

stats.pearsonr(x, y)
```
